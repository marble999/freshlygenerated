```python
import os
import sys
```

```python
replacements = {
    '\x92': '\'',
    '\x93': '"',
    '\x94': '"',
    '\x95': ' * ',
    '\x97': '--',
}
```

```python
articles = []
for fname in os.listdir('.'):
    if fname.endswith('.md'):
        print(fname)
        with open(fname, 'r', encoding = "ISO-8859-1") as f:
            md = f.read()
            for old, new in replacements.items():
                md = md.replace(old, new)
            if md.endswith('1. '):
                md = md[:-3]
        with open(fname, 'w') as f:
            f.write(md)
        articles.append(md)
```
